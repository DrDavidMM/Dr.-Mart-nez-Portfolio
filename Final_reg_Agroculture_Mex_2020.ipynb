```python
df_main4 = df_main3.sort_values("Sembrada", ascending = False)
df_main4
```

```python
fig = plt.figure(figsize = (15, 8))
plt.bar(df_main4["Nomestado"], df_main4["Sembrada"], color = 'tab:cyan')
plt.bar(df_main4["Nomestado"], df_main4["Cosechada"], color = 'blueviolet')
plt.xticks(rotation = 90)
plt.tick_params(axis = 'x', labelsize = 20)
plt.tick_params(axis ='y', labelsize = 20)
plt.title("Hectáreas sembradas y cosechadas en 2020", fontsize = 20)
plt.legend(['Sembrada', 'Cosechada'])

plt.show()
```

```python
fig = plt.figure(figsize = (15, 8))
plt.bar(df_main4["Nomestado"], df_main4["Biol_puro_necesario"], color = 'tab:cyan')
plt.xticks(rotation = 90)
plt.tick_params(axis = 'x', labelsize = 20)
plt.tick_params(axis = 'y', labelsize = 20)
plt.title("Demanda potencial de biol L", fontsize = 20)

plt.show()
```
